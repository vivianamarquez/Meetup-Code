# Connect Four
https://en.wikipedia.org/wiki/Connect_Four

The grid is 6 row by 7 columns, those being named from A to G.

You will receive a list of strings showing the order of the pieces which dropped in columns:

<img src="http://crowd-multilogue.com/Images/Codewars/KataConnectFourWinner.jpg">

The list may contain up to 42 moves and shows the order the players are playing.

The first player who connects four items of the same color is the winner.

You should return "Yellow", "Red" or "Draw" accordingly.


```python
import pandas as pd
import numpy as np
from tabulate import tabulate
```

```python
def check_winner(board):
    board2 = np.array(board)
    diagonals = [board2.diagonal(i) for i in range(-6+1,7)] + [board2[::-1,:].diagonal(i) for i in range(-6+1,7)]
    results = diagonals + [row for row in board2] + [board2[:,i] for i in range(7)]
    results = ["".join(result) for result in results]
    if 'RRRR' in results:
        return "Red"
    elif 'YYYY' in results:
        return "Yellow"
    else:
        None
```
